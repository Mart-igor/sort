```python
with open('input.txt', 'r', encoding='UTF-8') as file_in:
    n = int(file_in.readline())

result = 1  # Начальное значение для факториала

def rec(n):
    global result  # Указываем, что мы используем глобальную переменную result
    if n == 0:
        return result
    result *= n
    return rec(n - 1)

# Вызов функции и вывод результата
factorial = rec(n)
print(factorial)
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
    
def result(n, k):
    return factorial(n) / factorial(k) * factorial(n - k)

print(result(n, k))
```

```python
with open('input.txt', 'r', encoding='UTF-8') as file_in:
    n, k = tuple(file_in.read().split('\n'))
```

```python
def iterative(n, k):
    if k > n:
        return 0
    if k == 0 or k == n:
        return 1
    k = min(k, (n - k))
    c = 1
    for i in range(k):
        c = c * (n - i) // (i + 1)
    return c
    
print(iterative(n, k))
```

```python
def combinations(n, k):
    # Создаем двумерный массив dp размером (n + 1) x (k + 1)
    dp = [[0] * (k + 1) for _ in range(n + 1)]
    
    # Заполняем таблицу
    for i in range(n + 1):
        dp[i][0] = 1  # C(n, 0) = 1 для любого n
        for j in range(1, min(i, k) + 1):
            dp[i][j] = dp[i - 1][j - 1] + dp[i - 1][j]  # Рекуррентное соотношение

    return dp[n][k]

# Пример использования
n = 5
k = 2
result = combinations(n, k)
print(f"C({n}, {k}) = {result}")  # Вывод: C(5, 2) = 10
```

```python
import math 
def combinations_with_replacement(n, k):
    return math.factorial(n + k - 1) // (math.factorial(k) * math.factorial(n - 1))
```

```python
from scipy.special import comb
def combinations_with_replacement_scipy(n, k):
    return comb(n + k - 1, k, exact=True)
```
