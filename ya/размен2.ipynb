```python
def Change(money, table=None):
    if table is None:
        table = {}

    # Проверяем, было ли уже вычислено значение для данной суммы
    if money in table:
        return table[money]
    
    # Базовый случай
    if money == 0:
        table[money] = 0
        return table[money]
    
    # Инициализируем результат как бесконечность
    result = float('inf')
    
    # Перебираем доступные монеты
    for c in [1, 3, 4]:
        if c <= money:
            result = min(result, 1 + Change(money - c, table))
    
    # Сохраняем результат в таблице
    table[money] = result
    return table[money]

# Пример использования
money = 6
result = Change(money)
if result == float('inf'):
    print("Невозможно разменять данную сумму.")
else:
    print(f"Минимальное количество монет для размена {money}: {result}")
```

```python
def Change(money, coins, table):
    # Базовый случай: если сумма равна 0, нам не нужны монеты
    if money == 0:
        print(f"Change(0): Возвращаем 0 (базовый случай)")  # Выводим информацию о базовом случае
        return 0
    
    # Если сумма отрицательная, это невозможно, возвращаем бесконечность
    if money < 0:
        print(f"Change({money}): Возвращаем бесконечность (недопустимая сумма)")  # Выводим информацию о недопустимой сумме
        return float('inf')  # Невозможно разменять отрицательную сумму
    
    # Проверяем, есть ли уже сохраненное значение для этой суммы
    if money in table:
        print(f"Change({money}): Используем сохраненное значение {table[money]}")  # Выводим информацию о сохраненном значении
        return table[money]  # Возвращаем сохраненное значение
    
    # Инициализируем результат как бесконечность
    result = float('inf')  
    print(f"Change({money}): Начинаем вычисление")  # Выводим информацию о начале вычисления

    # Перебираем все доступные монеты
    for coin in coins:
        if coin <= money:  # Проверяем, можно ли использовать монету
            print(f"Change({money}): Пробуем монету {coin}")  # Выводим информацию о пробуемой монете
            # Рекурсивно вызываем Change для оставшейся суммы
            sub_result = Change(money - coin, coins, table)
            
            # Если sub_result не бесконечность, обновляем результат
            if sub_result != float('inf'):
                result = min(result, sub_result + 1)  # Добавляем 1 к количеству монет
                print(f"Change({money}): Обновляем результат до {result} (с учетом монеты {coin})")  # Выводим обновленный результат

    # Сохраняем результат в таблице
    table[money] = result  
    print(f"Change({money}): Сохраняем результат {result} в таблице")  # Выводим сохраненный результат
    return result  # Возвращаем минимальное количество монет для данной суммы

# Пример использования
coins = [1, 3, 4]  # Доступные монеты
table = {}  # Таблица для мемоизации
money = 6  # Сумма, которую нужно разменять

# Запускаем функцию
min_coins = Change(money, coins, table)
print(f"Минимальное количество монет для {money}: {min_coins}")
```
